```python
import pandas as pd
from ipyvizzu import Chart, Data, Config, Style

data_frame = pd.read_csv("https://raw.githubusercontent.com/vizzuhq/vizzu-workshops/main/2022-12-01-PyData-Global/data/LinkedIn_Poll.csv")
data = Data()
data.add_data_frame(data_frame)

display(data_frame)
```

```python
chart = Chart(width="100%", height="360px")

chart.animate(data)

chart.animate(
    Data.filter(
        "record.Group == 'Group 1'"
    ),
   Config({
            "x":{ "set": ["Vote[%]","Answer"], "ticks":True},
            "y":{ "set": "Group", "labels":False},
            "color": "Answer",
            "label": "Vote[%]",
            "title": "How often do you present your findings?"
        }),
    Style({
        'legend' : {'width' : '100%'},
    })
)
```

```python
chart.animate(
    Data.filter(None),
)
```

```python
chart.animate(
    Config({
        "split": True, 
        "title": "2 or more is the most popular answer in every group"
    }),
    Style({
      # 'title' : {'fontSize' : '1.8em'},
    })
)
```

```python
chart.animate(
    Config({
        "x": ["Vote count","Answer"], 
        "label": "Vote count", 
        "title": "61% of the votes came from one group"
    }),
)
```

```python
chart.animate(
    Config({ 
        "x": "Answer", 
        "y": "Vote count",
        "split": False, 
        "legend": "color",
        "title": "More than 700 people voted"
    })
)
```

```python
chart.animate(
    Config({ 
        "x": ["Total[%]","Answer"], 
        "y": None, 
        "label":"Total[%]"
    })
)
```

```python
chart.animate(
    Config({ 
        "coordSystem": "polar", 
        "title":"~68% presents at least once per month"
    })
)
```
